#Ejercicio IVA

Solicite el precio de 5 productos y para cada producto generele el IVA del 19% :
*   Generar el valor base de todos los productos.
*   Generar el valor total del IVA
*   Generar el valor total de la factura





```python
products=[float(input("Enter the value of the products: ")) for i in range(5)]
value = []
for i in range(len(products)):
  print("\nProduct value exclusive of tax ",i+1,": {} \nValue of tax : {} \nTotal : {} \n"
       .format(products[i],float(products[i])*0.19,float(products[i])*0.19+float(products[i])))
  value.append(float(products[i])*0.19+float(products[i]))
print(f"Total invoice amount : {sum(value)}")
```

#Ejercicio notas de estudiantes

Genere de forma aleatoria las notas entre 0.0 y 5.0 de 40 estudiantes de una asignatura, con esa informaci칩n calcular:


*   Cuantos estudiantes perdieron (menor o igual a 2.9).
*   Cuantos pasaron (3.0 o m치s).
*   Cual fue la nota promedio entre los 40 estudiantes.


```python
import random
notes = [random.randrange(0, 50)/10 for i in range(40)]
lost,passed = 0,0
for j in range(len(notes)):
  if notes[j] <= 2.9: lost +=1 
  else: passed += 1
print("The total number of students who passed is: {}\nThe total number of students who lost is: {}\nThe average is: {}"
      .format(passed,lost,sum(notes)/40))  
```
