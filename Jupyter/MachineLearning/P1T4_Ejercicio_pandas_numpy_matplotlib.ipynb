#Ejercicio de USUARIOS Y TIEMPO DE ACTIVIDAD

##Carga de datos USUARIOS Y TIEMPO DE ACTIVIDAD 


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 

df = pd.read_csv("/content/drive/MyDrive/IA/DataScience/user_timestamp.csv")
df.head(10)
```

##Id del usuario con mayor entradas al sistema

```python
#inplace=True : no hay necesidad de crear un objeto nuevo ==> df.sort_values('user_id',inplace=True)
#Ordenando los datos de forma ascendente, de forma descendente es con la propiedad ==> ascending=False  df.sort_values('user_id',ascending=False)
#value_counts devuelve el total de un valor repetido 

df = pd.read_csv("https://raw.githubusercontent.com/luisFernandoCastellanosG/Machine_learning/master/DataScience/DataSet's/user_timestamp.csv")

user = dict(df.value_counts('user_id'))
print("El id del usuario con mayor concurrencia en la web es: {}\nCon un total de entradas de : {} ".format(list(user)[0],max(user.values())))
#Menor concurrencia
print("El id del usuario con mayor concurrencia en la web es: {}\nCon un total de entradas de : {} ".format(list(user)[-1],min(user.values())))
#Grafica 
id = str(list(user)[0]), str(list(user)[-1])
values = max(user.values()),min(user.values())
plt.bar(id,values,color="orange")
plt.show()
```

##Cantidad de tiempo maximo y minimo

```python
df = pd.read_csv("https://raw.githubusercontent.com/luisFernandoCastellanosG/Machine_learning/master/DataScience/DataSet's/user_timestamp.csv")

time = dict(df.value_counts('timestamp'))

print("Total",list(time)[0]," max ", max(time.values()), "min",min(time.values()))
#Aparece 12 horas ya que con el metodo values_counts suma cada una de las horas repetidas 

#Grafica 
timestamp = str(list(time)[0]), str(list(time)[-1])
values = max(time.values()),min(time.values())
plt.barh(timestamp,values)
plt.show()

```

##Monto mas alto de un prestamo,total de prestamos, el prestamos mas realizado y el menos,prestamos inferiores a 10000
